```python
import issues
import sqlite3

conn = sqlite3.connect('./magazine_features.db')
cur = conn.cursor()
rows = cur.execute('select * from issue_html where magazine = "nymag"')
values = []
for row in rows:
    values.append(row)
```

```python
nymag_issues = [issues.NYMagIssue(row[-1], row[-2], row[-3]) for row in values]
```

```python
nymag_issues[1].feature_authors
```

```python
nymag_issues[5].url
```

```python
authors = []
for i, issue in enumerate(nymag_issues):
    try:
        cv = issue.cover_author
        authors.append([('nymag', issue.date, cv)])
    except:
        authors.append([('nymag', issue.date, None)])
    try:
        feature_authors = issue.feature_authors
        for author in feature_authors:
            authors.append([('nymag', issue.date, author)])
    except:
        print(i)
```

```python
authors_insert = [author[0] for author in authors]
cur.executemany('INSERT INTO authors VALUES (?,?,?)', authors_insert)
conn.commit()
```
